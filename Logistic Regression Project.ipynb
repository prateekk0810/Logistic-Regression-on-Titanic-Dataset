```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
```

```python
titanic_train=pd.read_csv('train titanic.csv',delimiter=',')
titanic_train
titanic_test=pd.read_csv('test titanic.csv',delimiter=',')
titanic_test
```

```python
titanic_train.drop(['Name','Ticket','Cabin'], axis=1, inplace=True)
titanic_test.drop(['Name','Ticket','Cabin'], axis=1, inplace=True)
```

```python
titanic_train
```

```python
titanic_train.Embarked.unique()

titanic_train.loc[titanic_train['Embarked']== 'S', ['Embarked']] = 0
titanic_train.loc[titanic_train['Embarked'] =='C', ['Embarked']] = 1
titanic_train.loc[titanic_train['Embarked'] =='Q', ['Embarked']] = 2

titanic_test.loc[titanic_test['Embarked'] =='S', ['Embarked']] = 0
titanic_test.loc[titanic_test['Embarked'] =='C', ['Embarked']] = 1
titanic_test.loc[titanic_test['Embarked'] =='Q', ['Embarked']] = 2
```

```python
titanic_train
```

```python
titanic_train.loc[titanic_train['Sex'] =='female', ['Sex']] = 0
titanic_train.loc[titanic_train['Sex'] =='male', ['Sex']] = 1

titanic_test.loc[titanic_test['Sex'] =='female', ['Sex']] = 0
titanic_test.loc[titanic_test['Sex'] =='male', ['Sex']] = 1
```

```python
titanic_test.Sex.unique()
```

```python
titanic_train
```

```python
print(titanic_train.shape)
print(titanic_test.shape)
```

```python
titanic_train.dropna(axis=0, inplace=True)
```

```python
titanic_test.isnull().count()
```

```python
titanic_train
```

```python
titanic_train.dtypes
```

```python
titanic_test.Embarked.mode()
```

```python
titanic_test.Age.fillna(titanic_test.Age.median(),inplace=True)
titanic_test.Embarked.fillna(titanic_test.Embarked.mode()[0],inplace=True)
```

```python
titanic_train['Sex']=titanic_train['Sex'].astype(str).astype(int)
titanic_train['Embarked']=titanic_train['Embarked'].astype(str).astype(int)

titanic_test['Sex']=titanic_test['Sex'].astype(str).astype(int)
titanic_test['Embarked']=titanic_test['Embarked'].astype(str).astype(int)
```

```python
titanic_train.dtypes
```

```python
titanic_train.Age.unique()
```

```python
x_train=titanic_train.to_numpy()[:,:-1]
y_train=titanic_train.to_numpy()[:,-1]
x_test=titanic_test.to_numpy()
print(x_train.shape)
print(y_train.shape)
```

```python
alg1=linear_model.LogisticRegression(C=10000,max_iter=10000,solver='liblinear')
alg1.fit(x_train,y_train)
```

```python
test_predictions=alg1.predict(x_test)
```

```python
np.savetxt('output_file.csv',test_predictions ,fmt='%0.2f')
```

```python
alg1.score(x_train, y_train)
```
